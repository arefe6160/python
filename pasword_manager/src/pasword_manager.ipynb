```python
from abc import ABC, abstractmethod

class paswordgenerator(ABC):
    def __init__(self,length):
        self.length=length
    @abstractmethod
    def generate(self):
        pass 
```

```python
class randompaswordgenerator(paswordgenerator):
    def __init__(self,length,include_numbers,include_symbols):
        self.length=length
        self.include_numbers=include_numbers
        self.include_symbols=include_symbols
    def generate(self):
        password = ''.join(random.choices(string.ascii_letters + string.digits + string.punctuation, k=self.length))
        return password
```

```python
word_list = [
    'Alice',
    'Bob',
    'Charlie',
    'David',
]
class memorablepaswordgenerator(paswordgenerator):
    def __init__(
            self,numofword: int = 4 ,
            seprator: str ='-',
            capitalized: bool =False,
            vocabulary: list =None
    ):
        if vocabulary is None:
            self.vocabulary = ['Alice','Bob','Charlie','David']

        self.capitalized = capitalized
        self.seprator = seprator
        self.numofword = numofword
    def generate(self):
        password = [random.choice(self.vocabulary)for _ in range (self.numofword)]
        if self.capitalized:
            password = [word.upper() if random.choice((True,False)) else word.lower() for word in password]
        return self.seprator.join(password)
```
