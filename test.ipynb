```python
import requests
```

```python
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
        }

requests.get("https://www.google.com",headers=headers)
```

```python
def check_website(url):
    try:
        response = requests.get(url, headers=headers,timeout=5)
        if response.status_code == 200:
            return True
        else:
            return False
    except Exception as e:
        return False
        
```

```python
website_to_check =[
     "https://www.github.com",
     "https://www.google.com",
     "https://www.stackoverflow.com",
     "https://www.twitter.com",
     "https://www.python.org",
     "https://www.ub.com"]*10
```

```python
for url in website_to_check:
    print(check_website(url))
```

```python
import threading
```

```python
check_website("https://www.google.com")
```

```python
thereds = []
for url in website_to_check:
    thread = threading.Thread(
        target=check_website, 
        args=(url,))
    thereds.append(thread)
    thread.start()
```

```python
for thread in thereds:
    thread.join()
```

```python
import concurrent
with concurrent.futures.ThreadPoolExecutor(max_workers=50) as executor:
    results = list(executor.map(check_website, website_to_check))   
```
